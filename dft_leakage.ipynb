```
import numpy as np
import matplotlib.pyplot as plt

N         = 128
omega     = 2*np.pi/N*25.3
k         = np.linspace(20, 30, 512, endpoint=True)
X         = np.sin( (omega-2*np.pi/N*k)*N/2 )/np.sin((omega-2*np.pi/N*k)/2)

k2        = np.arange(20,31)
X2        = np.sin( (omega-2*np.pi/N*k2)*N/2 )/np.sin((omega-2*np.pi/N*k2)/2)

plt.figure(figsize=(10,7))
plt.plot(k,X,'--')
plt.stem(k2,X2,linefmt='red')
plt.xlabel('k')
plt.ylabel('X(k)')
plt.grid()
```
